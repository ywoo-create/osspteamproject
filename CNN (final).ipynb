```python

# [0. н•„мҡ”н•ң лқјмқҙлёҢлҹ¬лҰ¬ м„Өм№ҳ л°Ҹ мһ„нҸ¬нҠё]
!pip install noisereduce -q  # л…ёмқҙмҰҲ м ңкұ° лқјмқҙлёҢлҹ¬лҰ¬ м„Өм№ҳ

import os
import random
import shutil
import numpy as np
import librosa
import soundfile as sf
import cv2
import matplotlib.pyplot as plt
import noisereduce as nr  # м „мІҳлҰ¬ л…ёнҠёл¶Ғмқҳ л…ёмқҙмҰҲ м ңкұ° лқјмқҙлёҢлҹ¬лҰ¬ м¶”к°Җ
from google.colab import drive
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import layers, models


# [1. кІҪлЎң л°Ҹ м„Өм • к°’]

DATASET_PATH = '/content/drive/MyDrive/мҳӨл””мҳӨнҢҢмқјwav'
classes = ['0_indoor_alarms', '1_outdoor_warnings', '2_emergency_alarms']
SAMPLE_RATE = 22050  # н‘ңмӨҖ мғҳн”Ңл§Ғ л ҲмқҙнҠё нҶөмқј


# [1.5. м—…лЎңл“ңлҗң л…ёнҠёл¶Ғмқҳ м „мІҳлҰ¬(Preprocessing) нҢҢмқҙн”„лқјмқё м •мқҳ]

def reduce_noise(y, sr=SAMPLE_RATE):
    """мҠӨнҺҷнҠёлҹҙ кІҢмқҙнҢ… кё°л°ҳ л…ёмқҙмҰҲ м ңкұ°. мһЎмқҢ кө¬к°„мқ„ л”°лЎң м§Җм •н•ҳм§Җ м•Ҡм•„лҸ„
    мҳӨл””мҳӨ м „мІҙм—җм„ң мһЎмқҢ н”„лЎңнҢҢмқјмқ„ нҶөкі„м ҒмңјлЎң м¶”м •н•ҙ м ңкұ°н•©лӢҲлӢӨ."""
    try:
        y_denoised = nr.reduce_noise(y=y, sr=sr, stationary=False)
    except Exception as e:
        print(f"  [кІҪкі ] л…ёмқҙмҰҲ м ңкұ° мӢӨнҢЁ, мӣҗліё мӮ¬мҡ©: {e}")
        y_denoised = y
    return y_denoised

def trim_silence(y, top_db=30):
    """м•һл’Өм—җ л¶ҷмқҖ л¬ҙмқҢ(лҳҗлҠ” кұ°мқҳ л¬ҙмқҢмқё) кө¬к°„мқ„ м ңкұ°н•©лӢҲлӢӨ.
    top_dbк°Җ лӮ®мқ„мҲҳлЎқ лҚ” мһ‘мқҖ мҶҢлҰ¬лҸ„ 'мҶҢлҰ¬ мһҲмқҢ'мңјлЎң мқёмӢқн•ҙ лҚң мһҗлҰ…лӢҲлӢӨ."""
    y_trimmed, _ = librosa.effects.trim(y, top_db=top_db)
    if len(y_trimmed) == 0:
        return y  # м „л¶Җ мһҳл ӨлӮҳк°ҖлҠ” кІҪмҡ°лҘј л°©м§Җн•ҳлҠ” м•Ҳм „мһҘм№ҳ
    return y_trimmed

def peak_normalize(y, target_peak=0.95):
    """н”јнҒ¬ кё°мӨҖмңјлЎң мқҢлҹүмқ„ м •к·ңнҷ”н•ҙм„ң нҢҢмқј к°„ мқҢлҹү м°ЁмқҙлҘј нҶөмқјн•©лӢҲлӢӨ."""
    max_val = np.max(np.abs(y))
    if max_val > 0:
        y = y / max_val * target_peak
    return y

def preprocess_audio(y, sr=SAMPLE_RATE):
    """м „мІҳлҰ¬ нҢҢмқҙн”„лқјмқё: л…ёмқҙмҰҲ м ңкұ° -> л¬ҙмқҢ нҠёлҰ¬л°Қ -> м •к·ңнҷ”"""
    y = reduce_noise(y, sr)
    y = trim_silence(y)
    y = peak_normalize(y)
    return y

# [2. мҳӨл””мҳӨ к°ңлі„ мҰқк°• н•ЁмҲҳ м •мқҳ]

def time_shift(y, sr=SAMPLE_RATE, shift_max=0.3):
    shift_amt = int(sr * random.uniform(-shift_max, shift_max))
    return np.roll(y, shift_amt)

def pitch_shift(y, sr=SAMPLE_RATE, n_steps_range=(-3, 3)):
    n_steps = random.uniform(*n_steps_range)
    return librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)

def time_stretch(y, rate_range=(0.85, 1.15)):
    rate = random.uniform(*rate_range)
    return librosa.effects.time_stretch(y, rate=rate)

def add_noise(y, noise_factor_range=(0.002, 0.01)):
    noise_factor = random.uniform(*noise_factor_range)
    noise = np.random.randn(len(y))
    return y + noise_factor * noise

def change_volume(y, gain_range=(0.6, 1.4)):
    gain = random.uniform(*gain_range)
    y_new = y * gain
    max_val = np.max(np.abs(y_new))
    if max_val > 1.0:
        y_new = y_new / max_val
    return y_new

def time_mask(y, sr=SAMPLE_RATE, mask_max_ratio=0.1):
    y_new = y.copy()
    mask_len = int(len(y) * random.uniform(0.02, mask_max_ratio))
    if mask_len <= 0 or len(y) <= mask_len:
        return y_new
    start = random.randint(0, len(y) - mask_len)
    y_new[start:start + mask_len] = 0.0
    return y_new

# лҰ¬мҠӨнҠё нҳ•нғңлЎң кҙҖлҰ¬н•ҳм—¬ л°ҳліөл¬ёмңјлЎң мІҳлҰ¬н•  мҲҳ мһҲлҸ„лЎқ кө¬м„ұ
AUGMENTATIONS = [
    time_shift,
    pitch_shift,
    time_stretch,
    add_noise,
    change_volume,
    time_mask
]


# [3. л©ң-мҠӨнҺҷнҠёлЎңк·ёлһЁ м¶”м¶ң л°Ҹ нҶөн•© мҰқк°• н•ЁмҲҳ (м „мІҳлҰ¬ нҢҢмқҙн”„лқјмқё м—°лҸҷ)]

def extract_mel_spectrogram(file_path, augment=False):
    try:
        # нҢҢмқј лЎңл“ң (sr=22050, лӘЁл…ёнҶӨ м Ғмҡ©)
        y, sr = librosa.load(file_path, sr=SAMPLE_RATE, mono=True, duration=3.0)

        # [мҲҳм •] нҢҢмқј лЎңл“ң м§Ғнӣ„ л…ёнҠёл¶Ғмқҳ м „мІҳлҰ¬ нҢҢмқҙн”„лқјмқё(л…ёмқҙмҰҲ м ңкұ°, нҠёлҰ¬л°Қ, м •к·ңнҷ”) м Ғмҡ©
        y = preprocess_audio(y, sr=sr)
    except Exception as e:
        print(f"  [м—җлҹ¬] {os.path.basename(file_path)} лЎңл“ң л°Ҹ м „мІҳлҰ¬ мӢӨнҢЁ: {e}")
        return []

    features = []

    # 1. м „мІҳлҰ¬лҗң мӣҗліё лҚ°мқҙн„°лЎң л©ң-мҠӨнҺҷнҠёлЎңк·ёлһЁ мғқм„ұ
    mel = librosa.feature.melspectrogram(y=y, sr=sr)
    features.append(cv2.resize(librosa.power_to_db(mel, ref=np.max), (64, 64)))

    # 2. мҰқк°• мҳөм…ҳмқҙ мјңм§„ кІҪмҡ° (нӣҲл Ё лҚ°мқҙн„°м…Ӣ кө¬м¶• мӢңм—җл§Ң мӢӨн–ү)
    if augment:
        for aug_func in AUGMENTATIONS:
            try:
                y_aug = aug_func(y)
                # мҰқк°• кё°лІ• м Ғмҡ© нӣ„ нҒҙлҰ¬н•‘ л°©м§ҖлҘј мң„н•ҙ лӢӨмӢң н”јнҒ¬ м •к·ңнҷ” мҲҳн–ү
                y_aug = peak_normalize(y_aug)

                mel_aug = librosa.feature.melspectrogram(y=y_aug, sr=sr)
                features.append(cv2.resize(librosa.power_to_db(mel_aug, ref=np.max), (64, 64)))
            except Exception as e:
                # нҠ№м • мҰқк°• кё°лІ• мӢӨнҢЁ мӢң мҠӨнӮөн•ҳкі  лӢӨмқҢ мҰқк°• м§„н–ү
                continue

    return features


# [4. лҚ°мқҙн„°м…Ӣ нҢҢмқј кІҪлЎң мҲҳм§‘]

print("\nл“ңлқјмқҙлёҢм—җм„ң нҢҢмқј лӘ©лЎқмқ„ мҲҳм§‘н•ҳлҠ” мӨ‘...")
file_paths = []
file_labels = []

for idx, class_name in enumerate(classes):
    class_dir = os.path.join(DATASET_PATH, class_name)
    if not os.path.exists(class_dir):
        print(f"[кІҪкі ] {class_name} нҸҙлҚ”лҘј м°ҫмқ„ мҲҳ м—ҶмҠөлӢҲлӢӨ. кІҪлЎңлҘј нҷ•мқён•ҳм„ёмҡ”.")
        continue

    for file in os.listdir(class_dir):
        if file.lower().endswith('.wav'):
            file_paths.append(os.path.join(class_dir, file))
            file_labels.append(idx)

print(f"мҙқ л°ңкІ¬лҗң WAV мӣҗліё нҢҢмқј мҲҳ: {len(file_paths)}к°ң")


# [5. лҚ°мқҙн„° лҲ„мҲҳ л°©м§ҖлҘј мң„н•ң Train / Test л¶„н•  (8:2)]

X_train_paths, X_test_paths, y_train_paths, y_test_paths = train_test_split(
    file_paths, file_labels, test_size=0.2, random_state=42, stratify=file_labels
)


# [6. нҠ№м§• м¶”м¶ң л°Ҹ лҚ°мқҙн„°м…Ӣ л№Ңл“ң (нӣҲл Ём…ӢмқҖ 1к°ң нҢҢмқјлӢ№ мҙқ 7к°ңлЎң мҰқк°•)]

print("\n л©ң-мҠӨнҺҷнҠёлЎңк·ёлһЁ нҠ№м§• м¶”м¶ң л°Ҹ 6мў… ліөн•© мҰқк°• м§„н–ү мӨ‘")
X_train, y_train = [], []
X_test, y_test = [], []

# нӣҲл Ё лҚ°мқҙн„°м…Ӣ кө¬м¶• (augment=TrueлЎң 6мў… кё°лІ• м „мӣҗ м Ғмҡ© -> мӣҗліё нҸ¬н•Ё мҙқ 7л°° мҰқк°Җ)
for path, label in zip(X_train_paths, y_train_paths):
    feats = extract_mel_spectrogram(path, augment=True)
    for f in feats:
        X_train.append(f)
        y_train.append(label)

# н…ҢмҠӨнҠё лҚ°мқҙн„°м…Ӣ кө¬м¶• (мҲңмҲҳ мӣҗліё лҚ°мқҙн„°лЎңл§Ң нҸүк°Җн•ҙм•ј н•ҳлҜҖлЎң augment=False)
for path, label in zip(X_test_paths, y_test_paths):
    feats = extract_mel_spectrogram(path, augment=False)
    for f in feats:
        X_test.append(f)
        y_test.append(label)


# [7. CNN мһ…л Ҙмқ„ мң„н•ң м°Ёмӣҗ мЎ°м • л°Ҹ л°°м—ҙ ліҖнҷҳ]

X_train = np.array(X_train)[..., np.newaxis] # (лҚ°мқҙн„°к°ңмҲҳ, 64, 64, 1)
y_train = np.array(y_train)
X_test = np.array(X_test)[..., np.newaxis]
y_test = np.array(y_test)

print(f"\n мөңмў… нӣҲл Ё лҚ°мқҙн„° нҳ•нғң (X_train): {X_train.shape}, л Ҳмқҙлё” к°ңмҲҳ: {len(y_train)}")
print(f" мөңмў… н…ҢмҠӨнҠё лҚ°мқҙн„° нҳ•нғң (X_test): {X_test.shape}, л Ҳмқҙлё” к°ңмҲҳ: {len(y_test)}")


# [8. CNN лӘЁлҚё м„Өкі„ л°Ҹ м»ҙнҢҢмқј]

model = models.Sequential([
    layers.Input(shape=(64, 64, 1)),
    layers.Rescaling(1.0/80.0, offset=1.0), # -80~0 лІ”мң„лҘј 0~1 лІ”мң„лЎң к°Җкіө

    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.4),
    layers.Dense(3, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()


# [9. лӘЁлҚё н•ҷмҠө мҲҳн–ү]

print("\n лӘЁлҚё н•ҷмҠө мӢңмһ‘")
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1
)


# [10. мөңмў… н…ҢмҠӨнҠём…Ӣ м„ұлҠҘ кІҖмҰқ]

print("\nн…ҢмҠӨнҠё лҚ°мқҙн„°лЎң мөңмў… м„ұлҠҘ нҸүк°Җ")
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"\nмөңмў… н…ҢмҠӨнҠё м •нҷ•лҸ„ (Test Accuracy): {test_acc * 100:.2f}%")


# [11. н•ҷмҠө м¶”мқҙ к·ёлһҳн”„ мӢңк°Ғнҷ”]

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Evolution')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Evolution')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.show()


# [12. лӘЁлҚё л°Ҹ к°ҖмӨ‘м№ҳ кІ°кіјл¬ј м ҖмһҘ л°Ҹ л“ңлқјмқҙлёҢ л°ұм—…]

model.save('my_audio_model.keras')
model.save_weights('my_audio_weights.weights.h5')

!cp my_audio_model.keras /content/drive/MyDrive/
!cp my_audio_weights.weights.h5 /content/drive/MyDrive/

print("лӘЁлҚё л°Ҹ нҢҢмқјмқҙ л“ңлқјмқҙлёҢм—җ м•Ҳм „н•ҳкІҢ л°ұм—…лҗҳм—ҲмҠөлӢҲлӢӨ.")
```
